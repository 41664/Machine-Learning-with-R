```r
# We notice that cy1 has a few values. 
#By default ggplot treats all values as continious variables 
# If we want a variable to be treated distinct we need to define it as a factor 
ggplot(mtcars, aes(x = factor(cyl), y = mpg, color=cyl, shape= factor(cyl))) +
  geom_point()
# color- on what colour the data representation aesthetic is based. Outline
# size- on what size is the aesthetic based 
# shape- only makes sense for categorical data. It defines different data in different shape. 
# shape- there are also 25 shapes available
# X- the x coordinate 
# Y - The Y coordinate
# Alpha- 0 is transparent and 1 is opaque 
# Fill- Fill colour. Inside
```

### Grammar of graphics 
1. Data- the data we need to plot
2. Aesthetics - The scales on which we need to plot
3. geom- the actual shape of the data we need to plot
4. Facets 
5. Stats
6. Coordinates 
7. Themes 
--------------------

we use several layers to plot the dataset. The first layer is always the data 

```r
# every layer can have it's own aesthetic locally
# geom_point() - adds the scatter points, all of them 
    # alpha- defines how alpha the colour is
    # show.legend
    # inherit.aes- Should inherit the aesthetics of parent?
# geom_smooth()- adds a smoothening line through the points 
    # se- display confidence interval bound? (the gray colour thing around line)
    # show.legend - show legend?
    # linetype- types of line. dotted or not
    # method Linear model. 
    #The method to draw a line e.g. "auto", "lm", "glm", "gam", "loess" or a function
ggplot(diamonds, aes(x = carat, y = price, color=clarity)) +
  geom_smooth(alpha=0.4, se=FALSE, show.legend= TRUE, linetype=5, method="auto")
```
